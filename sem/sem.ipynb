# connect to database

```python
import sqlite3
```

```python
connexion = sqlite3.connect("../db/vivino.db")
cursor = connexion.cursor()
```

## question 4

```python
cursor.execute("""
    SELECT
        DISTINCT(g.name), m.wines_count
    FROM
        countries as c
    JOIN most_used_grapes_per_country as m on c.code = m.country_code
    JOIN grapes as g on m.grape_id = g.id

               
""")
cursor.fetchall()
```

```python
cursor.execute("""
    ALTER TABLE grapes 
    ADD COLUMN wines_count integer; 
""")
```

```python
try:
    cursor.execute("""
        UPDATE grapes
        SET wines_count = (
        SELECT wines_count
        FROM most_used_grapes_per_country
        WHERE most_used_grapes_per_country.grape_id = grapes.id
    );
    """)
    connexion.commit()

    print("Update succesful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
        UPDATE wines
        SET grape_id = (
            SELECT most_used_grapes_per_country.grape_id
            FROM most_used_grapes_per_country
            JOIN regions ON most_used_grapes_per_country.country_code = regions.country_code
            WHERE regions.id = wines.region_id
        );
    """)
    connexion.commit()

    print("Update successful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
        ALTER TABLE vintages
        ADD COLUMN wineries TEXT;
    """)
    connexion.commit()

    print("Update successful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
        UPDATE vintages
        SET wineries = TRIM(
                          REPLACE(
                            name, 
                            (SELECT name FROM wines WHERE INSTR(vintages.name, wines.name) > 0), 
                            '')
                        );
    """)
    connexion.commit()

    print("Update successful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
        UPDATE vintages
        SET wineries = SUBSTR(wineries, 1, LENGTH(wineries) - 5)
        WHERE wineries IS NOT NULL;
    """)
    connexion.commit()

    print("Update successful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
        UPDATE vintages
SET wineries = CASE 
                    WHEN instr(vintages.name, wines.name) > 0 THEN TRIM(SUBSTR(vintages.name, 1, instr(vintages.name, wines.name) - 1))
                    ELSE wineries
                END
FROM wines
WHERE instr(vintages.name, wines.name) > 0;

    """)
    connexion.commit()

    print("Update successful!")
except Exception as e:
    print("Error:", e)
    connexion.rollback()  # Rollback changes if an error occurs
```

```python
try:
    cursor.execute("""
    UPDATE wines
    SET grape_id = COALESCE((
        SELECT grapes.id
        FROM grapes
        WHERE LOWER(wines.name) LIKE '%' || LOWER(grapes.name) || '%'
    ), (
        SELECT grapes.id
        FROM regions
        JOIN countries ON regions.country_code = countries.code
        JOIN most_used_grapes_per_country ON countries.code = most_used_grapes_per_country.country_code
        JOIN grapes ON most_used_grapes_per_country.grape_id = grapes.id
        WHERE wines.region_id = regions.id
    ), grape_id);
    """)
    connexion.commit()

    print("Grape IDs updated successfully.")
except Exception as e:
    print("Error:", e)
    connexion.rollback() 

```
